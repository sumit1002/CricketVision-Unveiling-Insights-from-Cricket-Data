```python
import pandas as pd
import json #for openening json files 
```

```python
with open ('json files/t20_wc_match_results.json') as f:
    data = json.load(f)
df_match= pd.DataFrame(data[0]['matchSummary'])
df_match.head()
```

```python
df_match.shape
```

```python
df_match.rename({'scorecard':'match_id'},axis=1,inplace= True)
df_match.head()
```

 Batting Summary // it does contain a single list so we need to convertb that 

```python
match_ids_dict={}
for index,row in df_match.iterrows():
    key1=row['team1']+' Vs '+row['team2']
    key2=row['team2']+' Vs '+row['team1']
    match_ids_dict[key1]=row["match_id"]
    match_ids_dict[key2]=row["match_id"]
match_ids_dict
```

```python
df_match.to_csv('t20_csv_files/match_summary.csv',index=False)
```

```python
with open ('json files/t20_wc_batting_summary.json') as f:
    data= json.load(f)
    all_records=[]
    for rec in data:
        all_records.extend(rec['battingSummary])
all_records
df_batting=pd.DataFrame(all_records)
df_batting.head()
```

extended is done because it conatains more than one
array of boeling records

create a new column for out and not out 


```python
df_batting["Out/Not Out"]= df_batting.dismissal.apply(lambda x: "out" if len(x)>0 else "not out" )
df_batting.head()
```

```python
df_batting.drop(columns=["dismissal"], inplace= True)
df_batting.head()
```

```python

df_batting["batsmanName"]= df_batting["batsmanName"].apply(lambda x: x.replace('â€','') )
df_batting.head(1)        
```

bring match id to this table for connecting in power bi 
so we did merge two teams above i for of dictionary with match id 
create a new coloumn match id in this 

```python
df_batting["match_id"]=df_batting["match"].map(match_ids_dict)
df_batting.head()
```

```python
df_batting.to_csv('t20_csv_files/batting_summary.csv',index=False)
```
